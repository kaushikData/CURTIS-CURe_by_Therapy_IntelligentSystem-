```python
df["word_len"] = df.Question.apply(lambda x:len(x.split()))
```

```python
(df.Question.str.find("?") != -1).value_counts()
```

```python
question_df = df[df.Question.str.find("?") != -1]
```

```python
question_df.head()
```

```python
question_df = question_df.reset_index()
```

```python
question_df = question_df.drop(['index'], axis = 1)
```

```python
question_df.head()
```

```python
question_df.rename(columns={'Question': 'questionText','Answer':'answerText'}, inplace=True)
```

```python
for i in range(question_df.shape[0]):
    # extract if a sentence in answer has sounds like or seems like
    if re.findall(r"([^.]*?(sounds like|seems like)[^.]*\.)",question_df.loc[i,"answerText"]):
        question_df.loc[i,"reflection"] = re.findall(r"([^.]*?(sounds like|seems like)[^.]*\.)",question_df.loc[i,"answerText"])[0][0] 
    # check if answer contains atleast 2 senetences, extract the second sentence if the first sentence is shorter (for example if "hello" is the first sentence) 
    # check oif hello or thank you is present in first sentence, then extract second sentence
    elif ((len(question_df.loc[i,"answerText"].split(".")[0].split()) <= 2 and len(question_df.loc[i,"answerText"].split(".")) >= 2) or (("hello" in question_df.loc[i,"answerText"].split(".")[0].lower() or "thank you" in question_df.loc[i,"answerText"].split(".")[0].lower()) and len(question_df.loc[i,"answerText"].split(".")) >= 2)):
        question_df.loc[i,"reflection"] = question_df.loc[i,"answerText"].split(".")[1].strip()
    # else just extract the first sentence as reflection
    else:
        question_df.loc[i,"reflection"] = question_df.loc[i,"answerText"].split(".")[0].strip()
```
